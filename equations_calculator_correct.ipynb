```python

def parse_linear(equations_str):
    b = []
    A = []
    for eq in equations_str:
        A.append([])
        eq_split = eq.split('=')
        b.append(float(eq_split[-1].strip()))
        sign = 1
        for x in eq_split[0].split(' '):
            if x != '+' and x != '-' and len(x) != 0:
                if '*' in x:
                    A[-1].append(sign*float(x.split('*')[0]))
                else:
                    try:
                        b[-1] -= sign*float(x)
                    except ValueError:
                        if '-' in x:
                            A[-1].append(-1.0*sign)
                        else:
                            A[-1].append(1.0*sign)
            elif x == '+':
                sign = 1
                
            elif x == '-':
                sign = -1
                
            else:
                continue
                
    return A, b
def parse_nonlinear(equation_str):
    eq_split = equation_str.split('=')
    d = [-1*float(eq_split[-1].strip())]
    coeffs = []

    sign = 1
    for elem in eq_split[0].split(' '):
        if elem != '+' and elem != '-' and len(elem) != 0:
            if '*' in elem:
                try:
                    coeff = elem.split('*')[0]
                    if '/' in coeff:
                        frac = coeff.split('/')
                        coeff = frac[0]/frac[1]
                        coeffs.append(sign*coeff)
                    else:
                        coeffs.append(sign*float(coeff))
                except ValueError:
                    continue
            else:
                try:
                    d.append(sign*float(elem))
                except ValueError:
                    if '-' in elem:
                        coeffs.append(-1.0*sign)
                    else:
                        coeffs.append(1.0*sign)
        elif elem == '+':
            sign = 1

        elif elem == '-':
            sign = -1

        else:
            continue
           
    coeffs.append(sum(d))
    return coeffs
def solve_slay(Am, bm):
    n = len(Am)
    A = Am.copy()
    b = bm.copy()
    for i in range(n):
        # 1
        scaler = 1.0 / A[i][i]
        for j in range(n): 
            A[i][j] *= scaler
        b[i] *= scaler

        # 2
        for k in range(n): 
            if k == i:
                continue
            scaler = A[k][i]
            for j in range(n):
                A[k][j] -= scaler * A[i][j]
            b[k] -= scaler * b[i]

    linear_solutions = b
    
    return linear_solutions

import math
def solve_3pol(coeffs):
    a, b, c, d = coeffs 

    f = ((3.0 * c / a) - ((b ** 2.0) / (a ** 2.0))) / 3.0
    g = (((2.0 * (b ** 3.0)) / (a ** 3.0)) - ((9.0 * b * c) / (a **2.0)) + (27.0 * d / a)) /27.0
    h = ((g ** 2.0) / 4.0 + (f ** 3.0) / 27.0)

    if f == 0 and g == 0 and h == 0:            
        if (d / a) >= 0:
            x = (d / (1.0 * a)) ** (1 / 3.0) * -1
        else:
            x = (-d / (1.0 * a)) ** (1 / 3.0)
        return [x, x, x]

    elif h <= 0:                               

        i = math.sqrt(((g ** 2.0) / 4.0) - h)
        j = i ** (1 / 3.0)
        k = math.acos(-(g / (2 * i)))
        L = j * -1
        M = math.cos(k / 3.0)
        N = math.sqrt(3) * math.sin(k / 3.0)
        P = (b / (3.0 * a)) * -1

        x1 = 2 * j * math.cos(k / 3.0) - (b / (3.0 * a))
        x2 = L * (M + N) + P
        x3 = L * (M - N) + P

        return [x1, x2, x3]

    elif h > 0:                                 
        R = -(g / 2.0) + math.sqrt(h) 
        if R >= 0:
            S = R ** (1 / 3.0)                  
        else:
            S = (-R) ** (1 / 3.0) * -1          
        T = -(g / 2.0) - math.sqrt(h)
        if T >= 0:
            U = (T ** (1 / 3.0))                
        else:
            U = ((-T) ** (1 / 3.0)) * -1

        x1 = (S + U) - (b / (3.0 * a))
        x2 = -(S + U) / 2 - (b / (3.0 * a)) + (S - U) * math.sqrt(3) * 0.5j
        x3 = -(S + U) / 2 - (b / (3.0 * a)) - (S - U) * math.sqrt(3) * 0.5j

        return [x1, x2, x3]
    
def solve_2pol(coeffs):
    a, b, c = coeffs
    D = b ** 2 - 4 * a * c
    print(D)
    if D < 0:
        return ['кореней нет']
    elif D == 0:
        x1 = -b / 2 * a
        return [x1]
    if D > 0:
        x1 = (-b + math.sqrt(D)) / (2 * a)
        x2 = (-b - math.sqrt(D)) / (2 * a)
        
        return [x1, x2]

def solve_bipol(coeffs):
    a, b, c = coeffs
    t = a**2
    D = b ** 2 - 4 * a * c
    print(D)
    if D == 0:
        x = -b / 2 * a
        x1 = x**(1/2)
        x2 = -(x)**(1/2)
        return [x1, x2]
    else:
        m = (-b + math.sqrt(D)) / (2 * a)
        n = (-b - math.sqrt(D)) / (2 * a)
    x1 = m**(1/2)
    x2 = -(m)**(1/2)
    x3 = n**(1/2)
    x4 = -(n)**(1/2)
    return [x1, x2, x3, x4]
def make_answer(lin_sol, nonlin_sol):
    answer = []
    for el in nonlin_sol:
        tmp = lin_sol.copy()
        tmp.append(el)
        answer.append(tmp)
    return answer
```
